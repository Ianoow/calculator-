Create a table `room` with columns`(room_id, room_name, room_floorno)` and another table `employees` with columns `(employee_id, employee_names, employee_address, employee_dob)`

```python
CREATE TABLE room (
    room_id int ,
    room_name VarChar(100) ,
    room_floorno int
);


CREATE TABLE employees (
    employee_id int,
    employee_names VARCHAR(150),
    employee_address VARCHAR(255),
    employee_dob DATE
);
```

Q6 Write a query to select all countries that are in continent Africa

```python
SELECT *FROM country
WHERE
    continent = 'Africa';
```

Q7 Write a query to select all cities with a `population` greater than 1,000,000 and are in `CountryCode` AFG

```python
SELECT *FROM city
WHERE
    countrycode = 'AFG'
    AND population > 1000000;
```

Q9 Using the country table, select all `countries` in Europe with a `LifeExpectancy` less than 40 years. LIMIT the records to 15 records only.

```python
SELECT *FROM country
WHERE continent = 'Europe'
  AND LifeExpectancy < 40
LIMIT 15;
```

 Q10 SELECT columns `Code`, `Continent`, `SurfaceArea`, `IndepYear` and `GNP` FROM the country table. Filter the records to only countries in North America and have a `surfaceArea` greater than 600000.

```python
SELECT Code, Continent, SurfaceArea, IndepYear, GNP
FROM country
WHERE Continent = 'North America'
  AND SurfaceArea > 600000;
```
