```python
def display_menu():

    print(" Contact Book ")
    print("1. Add Contact")
    print("2. View All Contacts")
    print("3. Search Contact")
    print("4. Exit")

def validate_email(email):
    return '@' in email and '.' in email.split('@')[-1]

def validate_phone(phone):

    return phone.replace('-', '').replace(' ', '').isdigit()

def add_contact():

    print(" Add New Contact")

    while True:
        name = input("Name: ").strip()
        if name:
            break
        print("Name cannot be empty!")

    while True:
        phone = input("Phone: ").strip()
        if validate_phone(phone):
            break
        print("Invalid phone number!")

    while True:
        email = input("Email: ").strip().lower()
        if validate_email(email):
            break
        print("Invalid email address!")

    with open("contacts.txt", "a") as file:
        file.write(f"{name},{phone},{email}\n")
    print(f"\nContact '{name}' added successfully!")

def load_contacts():

    try:
        with open("contacts.txt", "r") as file:
            return [line.strip().split(',') for line in file.readlines()]
    except FileNotFoundError:
        return []

def view_contacts():

    contacts = load_contacts()

    if not contacts:
        print("\nNo contacts found!")
        return

    print("\n--- All Contacts ")
    print(f"{'Name':<20}{'Phone':<15}{'Email':<25}")
    print("-" * 60)

    for name, phone, email in contacts:
        print(f"{name:}{phone:<15}{email:<25}")

def search_contact():
    """Search contacts by name"""
    contacts = load_contacts()

    if not contacts:
        print("\nNo contacts found!")
        return

    search_term = input("\nEnter name to search: ").strip().lower()
    results = [c for c in contacts if search_term in c[0].lower()]

    if not results:
        print("\nNo matching contacts found!")
        return

    print(f"\nFound {len(results)} matching contacts:")
    print(f"{'Name':<20}{'Phone':<15}{'Email':<25}")
    print("-" * 60)

    for name, phone, email in results:
        print(f"{name:<20}{phone:<15}{email:<25}")

def main():
    print("Welcome to Contact Book Manager!")

    while True:
        display_menu()
        choice = input("\nEnter your choice (1-4): ").strip()

        if choice == '1':
            add_contact()
        elif choice == '2':
            view_contacts()
        elif choice == '3':
            search_contact()
        elif choice == '4':
            print("\nGoodbye! Thank you for using Contact Book.")
            break
        else:
            print("Invalid choice! Please enter 1-4.")

        input("\nPress Enter to continue...")

if __name__ == "__main__":
    main()
```
