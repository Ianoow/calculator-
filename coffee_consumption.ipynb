```python
import pandas as pd

# Load your dataset (CSV example)
df = pd.read_csv("coffee.csv")

# Step 1: Create age groups (bins)
bins = [0, 20, 30, 40, 50, 60, 100]  # adjust based on your dataset
labels = ['<20', '20-29', '30-39', '40-49', '50-59', '60+']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Step 2: Calculate average coffee intake per age group
age_group_coffee = df.groupby('Age_Group')['Coffee_Intake'].mean().sort_values(ascending=False)

# Step 3: Display results
print("Average Coffee Intake by Age Group:")
print(age_group_coffee)

# Step 4 (Optional): Which age group consumes the most
max_group = age_group_coffee.idxmax()
max_value = age_group_coffee.max()
print(f"\nThe age group that consumes the most coffee is {max_group} with an average of {max_value:.2f} cups/day.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of sleep hours by coffee intake (e.g., using a histogram)
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Sleep_Hours', hue='Coffee_Intake', multiple='stack', bins=20)
plt.title('Distribution of Sleep Hours by Coffee Intake')
plt.xlabel('Sleep Hours')
plt.ylabel('Count')
plt.show()

# Distribution of sleep quality by coffee intake
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Sleep_Quality', hue='Coffee_Intake', palette='viridis')
plt.title('Distribution of Sleep Quality by Coffee Intake')
plt.xlabel('Sleep Quality')
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Create the DataFrame from the provided data


# Create coffee consumption categories
bins = [0, 1.5, 3.5, 10]
labels = ['Low (0-1.5 cups)', 'Moderate (1.6-3.5 cups)', 'High (3.6+ cups)']
df['Coffee_Category'] = pd.cut(df['Coffee_Intake'], bins=bins, labels=labels, include_lowest=True)

# Set up the visualization style
plt.style.use('seaborn-v0_8')
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Sleep Patterns by Coffee Consumption', fontsize=16, fontweight='bold')

# 1. Average sleep hours by coffee consumption category
sleep_by_coffee = df.groupby('Coffee_Category')['Sleep_Hours'].mean()
axes[0, 0].bar(sleep_by_coffee.index, sleep_by_coffee.values, color='steelblue', alpha=0.7)
axes[0, 0].set_title('Average Sleep Hours by Coffee Consumption')
axes[0, 0].set_ylabel('Average Sleep Hours')
axes[0, 0].tick_params(axis='x', rotation=45)

# Add value labels on bars
for i, v in enumerate(sleep_by_coffee.values):
    axes[0, 0].text(i, v + 0.05, f'{v:.2f}', ha='center', va='bottom')

# 2. Distribution of sleep hours by coffee consumption category
coffee_categories = df['Coffee_Category'].unique()
colors = ['lightblue', 'lightgreen', 'lightcoral']

for i, category in enumerate(coffee_categories):
    subset = df[df['Coffee_Category'] == category]
    axes[0, 1].scatter([i] * len(subset), subset['Sleep_Hours'], alpha=0.7,
                      color=colors[i], s=100, label=category)

axes[0, 1].set_title('Distribution of Sleep Hours by Coffee Consumption')
axes[0, 1].set_ylabel('Sleep Hours')
axes[0, 1].set_xticks(range(len(coffee_categories)))
axes[0, 1].set_xticklabels(coffee_categories, rotation=45)
axes[0, 1].legend()

# 3. Sleep quality distribution by coffee consumption category
sleep_quality_order = ['Poor', 'Fair', 'Good', 'Excellent']
df['Sleep_Quality'] = pd.Categorical(df['Sleep_Quality'], categories=sleep_quality_order, ordered=True)

# Create a cross-tabulation of sleep quality by coffee category
cross_tab = pd.crosstab(df['Coffee_Category'], df['Sleep_Quality'], normalize='index') * 100

# Plot stacked bar chart
cross_tab.plot(kind='bar', stacked=True, ax=axes[1, 0], color=['firebrick', 'darkorange', 'steelblue', 'seagreen'])
axes[1, 0].set_title('Sleep Quality Distribution by Coffee Consumption')
axes[1, 0].set_ylabel('Percentage (%)')
axes[1, 0].tick_params(axis='x', rotation=45)
axes[1, 0].legend(title='Sleep Quality', bbox_to_anchor=(1.05, 1), loc='upper left')

# 4. Relationship between coffee intake and sleep hours with sleep quality
sleep_quality_colors = {'Poor': 'firebrick', 'Fair': 'darkorange', 'Good': 'steelblue', 'Excellent': 'seagreen'}

for quality in sleep_quality_order:
    subset = df[df['Sleep_Quality'] == quality]
    if not subset.empty:
        axes[1, 1].scatter(subset['Coffee_Intake'], subset['Sleep_Hours'],
                          color=sleep_quality_colors[quality], alpha=0.7, s=100, label=quality)

axes[1, 1].set_title('Coffee Intake vs Sleep Hours (Colored by Sleep Quality)')
axes[1, 1].set_xlabel('Coffee Intake (cups/day)')
axes[1, 1].set_ylabel('Sleep Hours')
axes[1, 1].legend(title='Sleep Quality', bbox_to_anchor=(1.05, 1), loc='upper left')

# Add trendline
z = np.polyfit(df['Coffee_Intake'], df['Sleep_Hours'], 1)
p = np.poly1d(z)
axes[1, 1].plot(df['Coffee_Intake'], p(df['Coffee_Intake']), "r--", alpha=0.8)

# Calculate correlation
correlation = df['Coffee_Intake'].corr(df['Sleep_Hours'])
axes[1, 1].annotate(f'r = {correlation:.2f}', xy=(0.7, 0.9), xycoords='axes fraction')

plt.tight_layout()
plt.show()

# Additional statistics
print("Sleep Statistics by Coffee Consumption Category:")
for category in labels:
    subset = df[df['Coffee_Category'] == category]
    print(f"\n{category}:")
    print(f"  Average sleep hours: {subset['Sleep_Hours'].mean():.2f}")
    print(f"  Sleep quality distribution:")
    quality_counts = subset['Sleep_Quality'].value_counts(normalize=True) * 100
    for quality, percentage in quality_counts.items():
        print(f"    {quality}: {percentage:.1f}%")

# Correlation analysis
print(f"\nCorrelation between coffee intake and sleep hours: {correlation:.2f}")
```
